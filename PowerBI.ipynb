```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])

df.head()
```

```python
df.duplicated().sum()
```

```python
df[['Days', 'Hours']] = df['Waiting Time'].str.extract(r'(\d+) days, (\d+) hours')
df['Days'] = pd.to_numeric(df['Days'])
df['Hours'] = pd.to_numeric(df['Hours'])
```

```python
df['Total_Waiting_Days'] = df['Days'] + (df['Hours'] / 24)
```

```python
df.head()
```

```python
df.to_csv('cleaned_appointment_data.csv', index=False)
```

```python
df['Total_Waiting_Days'] = df['Total_Waiting_Days'].round(1)  # rounds to 1 decimal
```

```python
df.head()
```

```python
df.to_csv('final_cleaned_appointment_data.csv', index=False)
```

```python
# Group by Provider Role and calculate the average waiting time
average_waiting_time = df.groupby('Provider Role')['Total_Waiting_Days'].mean()

print(average_waiting_time.round(2))
```
